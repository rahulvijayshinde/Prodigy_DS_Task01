```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv("Metadata_country.csv")
df.head()
```

```python
df['Region'].value_counts()
```

```python
df_clean = df.dropna(subset=['Region'])
```

```python
region_counts = df_clean['Region'].value_counts()

plt.figure(figsize=(10,5))
region_counts.plot(kind='bar')
plt.title("Distribution of Countries by Region")
plt.xlabel("Region")
plt.ylabel("Number of Countries")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
income_counts = df['IncomeGroup'].value_counts()

plt.figure(figsize=(8,5))
income_counts.plot(kind='bar')
plt.title("Distribution of Countries by Income Group")
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.show()
```

```python
plt.savefig("region_distribution.png")
```

This bar chart represents the distribution of countries across different world regions.
It helps visualize which regions have the highest number of countries in the dataset.
Such visualization makes categorical data easier to interpret
